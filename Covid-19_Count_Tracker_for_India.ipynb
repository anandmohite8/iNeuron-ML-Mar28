```python
# do all the imports
import requests
import bs4
import tkinter as tk
import plyer
import time
import datetime
import threading
from bs4 import BeautifulSoup


#This function brings HTML data from host side which is Health Ministry of India Url:-
def get_html_data(url):
    data = requests.get(url)
    return data

#This function is used to scrape and parse HTML data the one we have received in our earlier function call:-
def get_corona_detail_of_india():
    url = "https://www.mohfw.gov.in/"
    html_data = get_html_data(url)
    bs=bs4.BeautifulSoup(html_data.text,'html.parser')
    info_div=bs.find("div",class_="site-stats-count").find_all("li",class_=["bg-blue","bg-green","bg-red","bg-orange"])#,"icon-dashboard"])
    
    all_details=""
    for block in info_div:
        text=block.find('span').get_text()
        count=block.find('strong').get_text()
        
        all_details=all_details+text+" : "+count+"\n"

    return all_details



# This function have been used to keep reloading the data from host site or keep refreshing the new content from the host site:-
def refresh():
    newdata=get_corona_detail_of_india()
    print("Refreshing ....")
    mainLabel['text']=newdata


    
# This function have been used to keep notifying the user at a given interval:-
def notify_me():
    while True:
        plyer.notification.notify(
        title='COVID 19 Cases within INDIA',
        message=get_corona_detail_of_india(),
        timeout=10000
        )
        time.sleep(30)
    
#This is a place where graphical user interface have been created:
root=tk.Tk()
root.geometry("900x800")
root.title("CoronaVirus-Live-Stats-For-India")
root.configure(background="light green")
f=("poppins", 25, "bold")


mainLabel=tk.Label(root,text=get_corona_detail_of_india(),font=f,bg='yellow')
mainLabel.pack()

reBtn=tk.Button(root, text="REFRESH", font=f, relief="solid", command =refresh)
reBtn.pack()

#This function have been created to generate a new thread :-
th1 = threading.Thread(target=notify_me)
th1.setDaemon(True)
th1.start()
                     
                     
root.mainloop()
```
